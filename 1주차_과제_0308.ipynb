###판다스 시작 - 파일을 DataFrame으로 로딩, 기본 API

  - pd.read_csv()
  : 파일 경로 지정, csv 파일을 DataFrame으로 로딩 가능
  - DataFrame.head()
  : DataFrame의 맨 앞에 있는 N개의 로우 반환 (Default 5개)
  - DataFrame.shape
  : DataFrame의 행과 열을 튜플 형태로 반환
  - DataFrame.info()
  : 총 데이터 건수와 데이터 타입, Null 건수를 알 수 있음
  - DataFrame.describe()
  : 칼럼별 숫자형 데이터 값의 n-percentile 분포도, 평균값, 최댓값, 최솟값을 나타냄, 숫자형 칼럼의 분포도만 조사하며 object 칼럼은 자동으로 제외
  - value_counts()
  : Series 해당 칼럼값의 유형과 건수 확인 가능
  
  <br>


###DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환

  - dict = df.to_dict('list')

###DataFrame 데이터 삭제
- drop() : axis값에 따라 특정 칼럼 또는 특정 행 드랍
  - labels : 삭제 원하는 칼럼 명
  
  - axis : axis0은 로우 방향 축, axis 1은 칼럼 방향 축

  - inplace : True면 자신의 DataFrame 데이터 삭제, default 값은 False (삭제된 결과 DataFrame 반환)

  <br>

새롭게 인덱스를 연속 숫자 형으로 할당
  - df.reset_index()

  <br>



###데이터 셀렉션 및 필터링
- [] : 행의 위치, 열의 위치, 슬라이싱 범위 등을 지정해 데이터 가져올 수 있음
- ix[] : 칼럼 명칭 기반 인덱싱, 현재는 판다스에 없음
- iloc[] : 위치 기반 인덱싱, 위치 인덱싱이 아닌 명칭을 입력하면 오류
- loc[] : 명칭 기반 인덱싱, 슬라이싱 기호를 적용하면 종료 값 -1이 아니라 종료 값까지 포함

<br>

###정렬, Aggregation 함수, GroupBy 적용
정렬
- sort_values() : DataFrame,Series 정렬
  - by : 특정 칼럼 입력하면 해당 칼럼으로 정렬 수행
  - ascending : True면 오름차순, False면 내림차순 (default는 True)
  - inplace : True면 호출한 DataFramd에 정렬 결과를 적용, False면 호출한 DataFrame 그대로 유지하며 정렬된 df를 결과로 반환 (default는 False)

Aggregation 함수

- DataFrame에서 바로 호출하면 모든 칼럼에 해당 aggregation 적용
  - count()
  - mean()
  - min(), max()
  - sum()

GroupBy
- groupby()
  - by : 칼럼을 입력하면 대상 칼럼으로 groupby

<br>


###결손 데이터 처리하기
- isna() : NAN 여부 확인
- fillna() : NAN 값을 다른 값으로 대체

<br>

ex) lambda_square = lambda x : x**2

-  map(lambda): 여러 개의 값을 입력 인자로 사용해야 할 경우

```python
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : 'child' if x<=15 else ('Adult' if x<=60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
```

```python
# 나이에 따라 세분화된 분류를 수행하는 함수 생성.
def get_category(age):
  cat = ''
  if age <= 5: cat = 'Baby'
  if age <= 12: cat = 'Child'
  if age <= 18: cat = 'Teenager'
  if age <= 25: cat = 'Student'
  if age <= 35: cat = 'Young Adult'
  if age <= 60: cat = 'Adult'
  else : cat = 'Elderly'

  return cat

# lambda 식에 위에서 생성한 get_category() 함수를 반환값으로 지정.
# get_category(X)는 입력값으로 'Age' 칼럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age','Age_cat']].head()
```
